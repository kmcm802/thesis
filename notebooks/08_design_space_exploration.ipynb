Uses the surrogate for dense design-space evaluation, feasibility mapping, and optimisation.

**Source:** `outputs/phase5/design_exploration.csv`, `optimisation_result.txt`

```python
from pathlib import Path
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from IPython.display import Image, display

ROOT = Path('..').resolve()
P5 = ROOT / 'outputs' / 'phase5'
```

## Contour Plots

```python
for fig_name in ['design_space_contours.png', 'efficiency_contour.png', 'feasible_region_D_R.png']:
    p = P5 / 'figures' / fig_name
    if p.exists():
        print(f'\n{fig_name}:')
        display(Image(filename=str(p), width=700))
```

## Feasibility Analysis

```python
de = pd.read_csv(P5 / 'design_exploration.csv')
print(f'Total design points evaluated: {len(de):,}')

# Apply constraints
feasible = de[
    (de['v_out_mean'] > 30) & 
    (de['efficiency'] > 0.85) & 
    (de['v_out_ripple'] < 1.0)
]
print(f'Feasible points: {len(feasible):,} / {len(de):,} = {len(feasible)/len(de)*100:.1f}%')

print(f'\nFeasible region parameter ranges:')
for col in ['D', 'V_in', 'R', 'f_sw']:
    print(f'  {col}: [{feasible[col].min():.3f}, {feasible[col].max():.3f}]')
```

## Surrogate-Based Optimisation

```python
print(open(P5 / 'optimisation_result.txt').read())
```

## Conclusion

- 49.8% of the 4D design space is feasible (V_out>30V, efficiency>85%, ripple<1V)

- The surrogate enables this exploration in 14 minutes vs 72.3 days for direct simulation
