# 01 — Simulator Validation (Phase 0)

Validates the analytical steady-state boost converter model against known physics.

**Tests:** Ideal-case, parasitic, monotonicity, convergence, timing.

**Source:** `outputs/phase0_validation.txt`, `outputs/phase0_timing.csv`

```python
from pathlib import Path
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

ROOT = Path('..').resolve()
print(open(ROOT / 'outputs' / 'phase0_validation.txt').read())
```

## Timing Distribution
100 simulations with random inputs across the design space.

```python
timing = pd.read_csv(ROOT / 'outputs' / 'phase0_timing.csv')
t = timing['wall_time_s'].values * 1000  # ms

fig, ax = plt.subplots(figsize=(8, 4))
ax.hist(t, bins=30, edgecolor='black', alpha=0.7)
ax.set_xlabel('Wall time per simulation (ms)')
ax.set_ylabel('Count')
ax.set_title(f'Phase 0 — Simulation timing (N=100, mean={t.mean():.4f} ms, std={t.std():.4f} ms)')
ax.axvline(t.mean(), color='red', linestyle='--', label=f'mean={t.mean():.4f} ms')
ax.legend()
plt.tight_layout()
plt.show()

print(f'Total: {t.sum()/1000:.6f} s')
print(f'Mean:  {t.mean():.4f} ms')
print(f'Std:   {t.std():.4f} ms')
```

## Conclusion

All 5 tests PASS. The analytical steady-state model is correct and evaluates in ~0.013 ms per point.
